```python
def process_videos(vid_range):
    for vid_idx in list(range(51,vid_range)):
        
        # Set Fail-Safes and initialize
        gui.PAUSE = 0.75 #set a pause time before each action
        gui.FAILSAFE = True # set a fail-safe method - moving the mouse to a screen corner will cause an exception, ending the script
        for i  in range(5):
            gui.hotkey('win', 'm')
            pass
        gui.hotkey('win', 'e') #open the file explorer
        time.sleep(2)
        gui.hotkey('win', 'up')
        gui.press('f4') #activate the navigation bar
        gui.hotkey('ctrl', 'a') #select the text
        gui.typewrite("L:\\2024.05.28") #type the experiment date folder
        gui.press('enter') #navigate to the experiment date folder
        gui.press('escape') #escape the navigation bar
        gui.PAUSE = 0.1 #set a pause time of 0.1 seconds between actions
        gui.hotkey('down') #initialize file highlighting
        gui.hotkey('up') #correct to first highlighted file

        i1=0 #initialize file counter
        while i1 < len(range(vid_idx)): #loop for video file selection
            gui.hotkey('down') #move selection down
            gui.hotkey('down') #move selection down (need two to account for the generated CSV)
            i1+=1 #increment index
        #gui.PAUSE(30)
        #gui.PAUSE(0.5)
        gui.press('enter') #load the highlighted video in FIJI

        gui.moveTo(288 , 20, 0.5) #select the file explorer window
        gui.click() #select the file explorer window
        gui.hotkey('ctrl', 'w') #close the file explorer
       
        time.sleep(15) #allow splash window to open
        
        gui.moveTo(1307 , 595, 0.5) #select the splash window
        gui.moveTo(1337 , 723, 0.5) #open with current settings
        gui.click() #open .nd2 file with current settings
        #time.sleep(90) #wait for the video to load
        #adjust contrast and properties
        gui.moveTo(1650, 630, 0.5) #move cursor to the video window
        time.sleep(3*60)
        gui.PAUSE = 0.5
        gui.click()
        gui.hotkey('shift', 'c') #open the FIJI brightness and contrast window

        gui.moveTo(1650, 630, 0.5) #move to somewhere
        gui.moveTo(842 , 452, 0.5) #move to the "Apply" button for contrast
        gui.click()
        gui.press('enter') # select the "OK" change contrast confirmation option

        gui.moveTo(1650, 630, 0.5) #move cursor to the video window
        gui.click()
        gui.hotkey('shift', 'p') #open the FIJI properties window
        gui.moveTo(980 , 462, 0.5) #move to the framerate dialog box
        gui.click()
        gui.hotkey('ctrl', 'a') #select the text in the framefate dialog box
        gui.press('backspace') #delete the text in the framerate dialog box
        gui.typewrite("1 sec") #type "1 sec" in the framerate dialog box
        gui.moveTo(1014 , 584, 0.5) #move to the "OK" button
        gui.click()

        # open 3DTRAX
        gui.moveTo(1245 , 34, 0.5) #move cursor to the video header
        time.sleep(0.5) #wait 0.5 seconds
        gui.dragTo(1300, 150, 1, button='left') #drag the window down to access the main menu

        gui.moveTo(1389, 15, 0.5)  #select the FIJI main page
        gui.click()
        gui.hotkey('win', 'up')
        
        gui.moveTo(606 , 46, 0.5)  #select the "3DTRAX" FIJI menu item
        gui.click()
        gui.moveTo(611, 73, 0.5) #select the 3DTRAX Analysis item
        gui.click()
        time.sleep(8)
        
        #open the calibration file
        gui.moveTo(1036 , 518, 0.5) #move cursor to the "Open" calibration file button in 3DTRAX
        gui.click()
        #gui.moveTo(965 , 642) #move cursor to the search bar
        #gui.click() #select the search bar
        gui.typewrite(r"L:\2024.05.28\calibration_003_29-May-2024_13-23-35(78Elements).3DTRAX.calibration") #input the calibration file name
        gui.press("enter") #import the calibration file into 3DTRAX

        #Select the video analysis method as "Sparse emitter"
        gui.moveTo(1117, 590, 0.5) #select analysis method for localization
        gui.click()
        gui.moveTo(1117, 617, 0.5) #set to sparse emitter
        gui.click()

        gui.moveTo(1037 , 632, 0.5) #refresh video source to currently open video file
        gui.click()

        gui.moveTo(1253, 782, 0.5) #run analysis

        gui.click() #RUN ANALYSIS
        time.sleep(22*60)
        gui.moveTo(1178, 730, 0.5) #save results
        gui.click()
        time.sleep(2)
        gui.moveTo(1897, 18, 0.5) #close FIJI
        gui.click()
        gui.sleep(1)  #wait for FIJI to close
        gui.moveTo(1004 , 364, 0.5) #don't save changes to video
        gui.click(1004 , 364) #click ""don't save" changes to video

        gui.sleep(2)  #wait for FIJI to close
```
