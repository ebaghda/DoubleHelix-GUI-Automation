```python
import time
import pyautogui as gui
```

```python
def process_videos(vid_range):
    for vid_idx in list(range(int(118/2),vid_range)):
        
        # Set Fail-Safes and initialize
        gui.PAUSE = 0.75 #set a pause time of 
        gui.FAILSAFE = True # set a fail-safe method - moving the mouse to a screen corner will cause an exception, ending the script
        for i  in range(5):
            #gui.hotkey('win', 'm')
            pass
        

        gui.hotkey('win', 'r')
        gui.typewrite("C:\Fiji.app\ImageJ-win64.exe")
        gui.press('enter')
        time.sleep(3)
        found=False
        while found is False:
            try:
                loc = gui.locateOnScreen("doubleHelixMenuScreenshot.png", grayscale=True)
                found=True
            except:
                time.sleep(3)
        gui.click(loc.left, loc.top-15)
        gui.hotkey('win', 'up')
        gui.hotkey('ctrl', 'o')
        gui.typewrite(r"I:\2024.06.05")
        gui.press('enter')
        found2 = False
        while found2 is False:
            try:
                loc = gui.locateOnScreen("FileIndexScreenshot.png", grayscale=True)
                found2=True
            except:
                time.sleep(2)
        gui.click(loc.left, loc.top)

        gui.PAUSE = 0.1 #set a pause time of 0.3
        i1=0 #initialize file counter
        while i1 < len(range(vid_idx)): #loop for video file selection
            gui.hotkey('down') #move selection down
            gui.hotkey('down') #move selection down (need two to account for the generated CSV)
            i1+=1 #increment index
        gui.pause = 1
        gui.press('enter') #load the highlighted video in FIJI
       
        time.sleep(2) #allow splash window to open
        while gui.pixelMatchesColor(1130, 647, (255, 255, 255), tolerance=10) is False:
            time.sleep(1)
            gui.move(-2, -2)
            gui.move(2, 2)
        gui.moveTo(650, 140, 0.5) #select the splash window
        gui.moveTo(1250, 630, 0.5) #open with current settings
        gui.click() #open .nd2 file with current settings
        #time.sleep(90) #wait for the video to load
        #adjust contrast and properties
        gui.moveTo(1650, 630, 0.5) #move cursor to the video window
        found = False
        while found is False:
            try:
                loc = gui.locateOnScreen("ScreenshotPlayButton.png", grayscale=True)
                gui.moveTo(loc.left+4, loc.top+4)
                gui.click()
                found=True
            except:
                time.sleep(10)
                gui.move(-1, -0)
                gui.move(1, 0)
                gui.hotkey('ctrl', '7')
                found1=False
                while found1 is False:
                    try:
                        loc1=gui.locateOnScreen("ScreenshotNoImagesOpenWarning.png", grayscale=True)
                        gui.moveTo(loc1.left+80, loc1.top+55)
                        gui.click()
                        found1=True
                    except:
                        time.sleep(1)
                        try:
                            gui.locateOnScreen("ScreenshotPlayButton.png", grayscale=True)
                            break
                        except:
                            time.sleep(1)
      
        gui.hotkey('shift', 'p') #open the FIJI properties window
        gui.moveTo(950, 420, 0.5) #move to the framerate dialog box
        gui.click()
        gui.hotkey('ctrl', 'a') #select the text in the framerate dialog box
        gui.press('backspace') #delete the text in the framerate dialog box
        gui.typewrite("1 sec") #type "1 sec" in the framerate dialog box
        gui.moveTo(1000, 520, 0.5) #move to the "OK" button
        gui.click()

        # open 3DTRAX
        #gui.moveTo(1300, 20, 0.5) #move cursor to the video header
        #time.sleep(0.5) #wait 0.5 seconds
        #gui.dragTo(1300, 150, 1, button='left') #drag the window down to access the main menu

        #gui.moveTo(1389, 15, 0.5)  #select the FIJI main page
        #gui.click()
        #gui.hotkey('win', 'up')
        
        gui.moveTo(488, 32, 0.5)  #select the "3DTRAX" FIJI menu item
        gui.click()
        gui.moveTo(498, 59, 0.5) #select the 3DTRAX Analysis item
        gui.click()
        while gui.pixelMatchesColor(770, 198, (9, 138, 239), tolerance=10) is False:
            time.sleep(3) #wait for the 3DTRAX page to open

        #open the calibration file
        gui.moveTo(1030, 500, 0.5) #move cursor to the "Open" calibration file button in 3DTRAX
        gui.click()
        gui.sleep(0.5)
        gui.typewrite(r"L:\2024.05.28\calibration_003_29-May-2024_13-23-35(78Elements).3DTRAX.calibration") #input the calibration file name
        gui.press("enter") #import the calibration file into 3DTRAX
        time.sleep(2)

        #Select the video
        gui.moveTo(1100,560, 0.5) #select analysis method for localization
        gui.click()
        gui.moveTo(1100, 580, 0.5) #set to sparse emitter
        gui.click()

        gui.moveTo(1020, 600, 0.5) #refresh video source to currently open video file
        gui.click()

        gui.moveTo(1190, 715, 0.5) #run analysis

        gui.click() #RUN ANALYSIS
        while gui.pixelMatchesColor(1155, 673, (215, 228, 241), tolerance=10) is False:
            time.sleep(1)
            gui.move(-1, 0)
            gui.move(1, 0)
            if gui.pixelMatchesColor(1058, 904, (208, 157, 0), tolerance=10) is True:
                break
        gui.moveTo(1150, 675, 0.5) #save results
        gui.click() #click "Save"
        time.sleep(3)
        gui.moveTo(1893, 14, 0.5) #close FIJI
        gui.click()
        gui.sleep(1)  #wait for FIJI to close
        gui.moveTo(980, 345, 0.5) #don't save changes to video
        gui.click()

        gui.sleep(2)  #wait for FIJI to close
```

```python
process_videos(116)
print("Completed running")
```
