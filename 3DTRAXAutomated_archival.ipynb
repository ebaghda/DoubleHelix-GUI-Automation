```python
import time
import pyautogui as gui
```

```python
def process_videos(vid_range):
    for vid_idx in range(vid_range):
        # Set Fail-Safes and initialize
        gui.PAUSE = 0.5
        gui.FAILSAFE = True
        for i in range(5):
            gui.hotkey('win', 'm')

        gui.moveTo(750, 750, 0.5)
        gui.click()
        gui.hotkey('win', 'e')
        gui.press('f4')
        time.sleep(1)
        gui.hotkey('ctrl', 'a')
        time.sleep(1)
        gui.typewrite("L:\\2024.05.28")
        gui.press('enter')
        gui.press('escape')
        gui.hotkey('down')
        gui.hotkey('up')

        i1=0
        while i1 < len(range(vid_idx)):
            gui.hotkey('down')
            gui.hotkey('down')
            i1+=1
        gui.press('enter') #open the video
        time.sleep(10)
        gui.hotkey('win', 'm')


        gui.moveTo(650, 140, 0.5)
        gui.click()
        gui.moveTo(1250, 630, 0.5)
        gui.click()
        time.sleep(120)

        #adjust contrast and properties
        gui.moveTo(1650, 630, 0.5)
        gui.click()
        gui.hotkey('shift', 'c')

        gui.moveTo(1650, 630, 0.5)
        gui.moveTo(800, 370, 0.5)


        gui.click()

        gui.moveTo(910, 325, 0.5)
        gui.click()

        gui.moveTo(1650, 630, 0.5)
        gui.click()
        gui.click()
        gui.hotkey('shift', 'p')
        gui.moveTo(950, 420, 0.5)
        gui.click()
        gui.hotkey('ctrl', 'a')
        gui.press('backspace')
        gui.typewrite("1 sec")
        gui.moveTo(1000, 520, 0.5)
        gui.click()

        # open 3DTRAX
        gui.moveTo(1300, 20, 0.5)
        time.sleep(2)
        gui.dragTo(1300, 150, 1, button='left')

        #open the calibration file
        gui.moveTo(1550, 40, 0.5)
        gui.click()
        gui.moveTo(1550, 60, 0.5)
        gui.click()
        time.sleep(4)

        gui.moveTo(1030, 500, 0.5)
        gui.click()

        gui.moveTo(1150, 390, 0.5)
        gui.click()
        gui.moveTo(700,500, 0.5)
        gui.click()
        gui.press('enter')

        gui.moveTo(1235,570, 0.5)
        gui.click(clicks=10, interval=0.25)
        gui.moveTo(735,570, 0.5)
        gui.click(clicks=2, interval=0.1)

        gui.moveTo(1235,599, 0.5)
        gui.click()
        gui.typewrite("2024.05.28")
        gui.press("enter")
        gui.moveTo(1235,599, 0.5)
        gui.click(clicks=2, interval=0.1)
        gui.typewrite("calibration_003_29-May-2024_13-23-35(78Elements).3DTRAX.calibration")
        gui.press("enter")

        #Select the video
        gui.moveTo(1100,560, 0.5) #select analysis method
        gui.click()
        gui.moveTo(1100, 580, 0.5) #set to sparse emitter
        gui.click()

        gui.moveTo(1020, 600, 0.5) #refresh video source
        gui.click()

        gui.moveTo(1190, 715, 0.5) #run analysis

        gui.click() #RUN ANALYSIS
        time.sleep(20*60) #WAIT 20 MINUTES

        gui.moveTo(1150, 675, 0.5) #save results
        gui.click()

        gui.moveTo(1840, 145, 0.5) #close video
        gui.click()
        gui.moveTo(980, 345, 0.5) #don't save changes to video
        gui.click()

        gui.moveTo(1150, 25, 0.5)
        gui.click()
        gui.moveTo(1150, 90, 0.5) #run garbage collector
        gui.click()
        time.sleep(2)

        gui.moveTo(1640, 25, 0.5)
        gui.click()
        gui.sleep(3)
```
